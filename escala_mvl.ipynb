```python
def verificar_rede_por_data(indice_data):

    coluna = df.columns[11 + indice_data]

    resultado = df[
        (df["Op"] == "MVL") &
        (df[coluna] == "REDE")
    ]

    print("=" * 50)
    print(f"REDES PARA {datas.iloc[indice_data]}")
    print("=" * 50)

    for _, linha in resultado.iterrows():

        print(
            f"{linha['Cidade']} | "
            f"{linha['Turno']} | "
            f"{linha['Colaborador']}"
        )

    print(f"\nTotal: {len(resultado)}")
```

```python
def consultar_rede(data):

    coluna = mapa_datas[data]

    resultado = df[
        (df["Op"] == "MVL") &
        (df[coluna] == "REDE")
    ]

    return resultado[
        ["Cidade", "Turno", "Colaborador"]
    ]
```

```python
consultar_rede("2026-06-15")
```

```python
def consultar_rede(data, operacao="MVL"):

    coluna = mapa_datas[data]

    resultado = df[
        (df["Op"] == operacao) &
        (df[coluna] == "REDE")
    ]

    return resultado[
        ["Cidade", "Turno", "Colaborador"]
    ]
```

```python
consultar_rede("2026-06-15")
```

```python
consultar_rede("2026-06-15", "MVL")
```

```python
def total_redes(data, operacao="MVL"):

    return len(
        consultar_rede(data, operacao)
    )
```

```python
total_redes("2026-06-15")
```
